```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline
from IPython.display import Audio
```

```python
N = 256;                 # length of test signals
num_freqs = 100;         # number of frequencies to test

# Generate vector of frequencies to test
omega = np.pi/8 + np.linspace(0,num_freqs-1,num_freqs)/num_freqs*np.pi/4;

S = np.zeros([N,num_freqs]);                        # matrix to hold FFT results

#rectangular window with no zero-padding 

for i in range(0,len(omega)):                       # loop through freq. vector
    s = np.sin(omega[i]*np.linspace(0,N-1,N));      # generate test sine wave
    win = signal.boxcar(N);                         # use rectangular window
    s = s*win;                                      # multiply input by window
    S[:,i] = np.square(np.abs(np.fft.fft(s)));      # generate magnitude of FFT
                                                    # and store as a column of S

plt.plot(S);                                        # plot all spectra on same graph
plt.title("rectangular with no zero padding")
```

```python

for i in range(0,len(omega)):                       # loop through freq. vector
    s = np.sin(omega[i]*np.linspace(0,N-1,N));      # generate test sine wave
    win = signal.hamming(N);                         # use rectangular window
    s = s*win;                                      # multiply input by window
    S[:,i] = np.square(np.abs(np.fft.fft(s)));      # generate magnitude of FFT
                                                    # and store as a column of S

plt.plot(S);                                        # plot all spectra on same graph
plt.title("hamming with no zero padding")
plt.show(); 
```

```python
ZeroPad = 256*4 ; 


# Generate vector of frequencies to test
omega = np.pi/8 + np.linspace(0,num_freqs-1,num_freqs)/num_freqs*np.pi/4;

S = np.zeros([ZeroPad,num_freqs]);                        # matrix to hold FFT results

# Create a zero-padded version of x
#xpad = [x, zeros(1,(K-1)*length(x))];

for i in range(0,len(omega)):                       # loop through freq. vector
    s = np.sin(omega[i]*np.linspace(0,N-1,N));      # generate test sine wave
    win = signal.boxcar(N);                         # use rectangular window
    s = s*win;                                      # multiply input by window
    S[:,i] = np.square((np.abs(np.fft.fft(s, ZeroPad))));      # generate magnitude of FFT
                                                    # and store as a column of S

plt.plot(S);                                        # plot all spectra on same graph
plt.title("Rectangular window with factor 4 zero padding")
```

```python
for i in range(0,len(omega)):                       # loop through freq. vector
    s = np.sin(omega[i]*np.linspace(0,N-1,N));      # generate test sine wave
    win = signal.hamming(N);                         # use rectangular window
    s = s*win;                                      # multiply input by window
    S[:,i] = np.square((np.abs(np.fft.fft(s, ZeroPad))));      # generate magnitude of FFT
                                                    # and store as a column of S

plt.plot(S);                                        # plot all spectra on same graph
plt.title("hamming with factor 4 zero padding")
```

```python
#Assignment 4
f1 = 2000
f2 = 2100
pi = np.pi
N = 256
sampling = 8000
n = np.linspace(0,N,N-1)/sampling
x1 = np.sin(2*pi*f1*n)
x2 = np.sin(2*pi*f2*n)
x = x1 + x2

f, t, Sxx = signal.spectrogram(x1,
                               fs = sampling*1.0,
                               nperseg = 64,
                               noverlap = 0)
plt.pcolormesh(t, f, Sxx)
plt.title("spectrogram of 2kHz with no overlap")
plt.xlabel("time/s")
plt.ylabel("frequency/Hz")

plt.figure()

f, t, Sxx = signal.spectrogram(x1,
                               fs = sampling*1.0,
                               nperseg = 64,
                               noverlap =32)
plt.pcolormesh(t, f, Sxx)
plt.title("spectrogram of 2kHz with 505 overlap")
plt.xlabel("time/s")
plt.ylabel("frequency/Hz")

plt.figure()

f, t, Sxx = signal.spectrogram(x2,
                               fs = sampling*1.0,
                               nperseg = 64,
                               noverlap = 0)
plt.pcolormesh(t, f, Sxx)
plt.title("spectrogram of 2kHz with no overlap")
plt.xlabel("time/s")
plt.ylabel("frequency/Hz")

plt.figure()

f, t, Sxx = signal.spectrogram(x2,
                               fs = sampling*1.0,
                               nperseg = 64,
                               noverlap =32)
plt.pcolormesh(t, f, Sxx,)
plt.title("spectrogram of 2kHz with 50% overlap")
plt.xlabel("time/s")
plt.ylabel("frequency/Hz")

print('''50% overlap shows spectrum over some of the data map
\nWithout doing overlap, you can see data over the entire map''')
```

```python
t = np.linspace(0,0.5,4001)
s = signal.chirp(t,1000,0.5,5000);    # Frequency-sweep signal

Audio(s,rate=8192)    # Default rate is 8192Hz
```

```python
f, t, Sxx = signal.spectrogram(s,
                               fs = 8192*1.0,
                               nperseg = 256,
                               noverlap =0)
plt.pcolormesh(t, f, Sxx)
plt.title("spectrogram frequency sweeper")
plt.xlabel("time/s")
plt.ylabel("frequency/Hz")

plt.figure()

f, t, Sxx = signal.spectrogram(s,
                               fs = 8192*1.0,
                               nperseg = 256,
                               noverlap =128)
plt.pcolormesh(t, f, Sxx)
plt.title("spectrogram of frequency sweeper with 50% OVERLAP")
plt.xlabel("time/s")
plt.ylabel("frequency/Hz")

print("At 0.4 second, the frequency has its peak")
```
