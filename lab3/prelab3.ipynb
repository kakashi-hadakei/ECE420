```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline
```

```python
#Problem1
'''
1)Hamming window and hanning window lower the amplitude of side lobes, while it has larger width of the bandlobe

2)The zero padding does increas frequency resolution as frequency change goes from 2pi/N to 2pi/(N+M)

3)DFT samples the value of DTFT over the interval 0 to 2pi

'''
```

```python
N = 256;                 # length of test signals
num_freqs = 100;         # number of frequencies to test

# Generate vector of frequencies to test
omega = np.pi/8 + np.linspace(0,num_freqs-1,num_freqs)/num_freqs*np.pi/4;

S = np.zeros([N,num_freqs]);                        # matrix to hold FFT results

for i in range(0,len(omega)):                       # loop through freq. vector
    s = np.sin(omega[i]*np.linspace(0,N-1,N));      # generate test sine wave
    win = signal.boxcar(N);                         # use rectangular window
    s = s*win;                                      # multiply input by window
    S[:,i] = np.square(np.abs(np.fft.fft(s)));      # generate magnitude of FFT
                                                    # and store as a column of S

plt.plot(S[:,3]);                                        # plot all spectra on same graph
```

```python
f1 = 2000
f2 = 2100
pi = np.pi
sampling = 8000
n = np.linspace(0,N,N-1)/sampling
x = np.sin(2*pi*f1*n)+np.sin(2*pi*f2*n)
size = len(x)
X = (np.fft.fft(x));
X = np.fft.fftshift(X)

w = 2*pi*(np.fft.fftfreq(size))
w = np.fft.fftshift(w)
analog = sampling * w /2/pi
plt.plot(n,x)
```

```python
plt.plot(analog,(np.absolute(X)))
plt.title("magnitude response")
plt.xlabel("analog frequency(Hz)")
plt.ylabel("Magnitude")

plt.figure()
plt.plot(analog,np.angle(X))
plt.title("phase response")
plt.xlabel("analog frequency")
plt.ylabel("Phase")
```
