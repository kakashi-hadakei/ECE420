```python
import numpy as np
import scipy
from scipy import signal,io
import matplotlib.pyplot as plt

print("module imported")
```

```python
sampling_freq = 30  # Hz
nyq = sampling_freq / 2

#50bpm to 140bpm
bpm_low = 50
bpm_high = 140
cutoff_low = 1.0*bpm_low/60
cutoff_high = 1.0*bpm_high/60

passband_left_one = cutoff_low-0.001
passband_left_two = cutoff_low+0.001
passband_right_one = cutoff_high-0.1
passband_right_two = cutoff_high+0.1

#print(passband_left_one,passband_left_two,passband_right_one,passband_right_two)

tap = 51

band = [0,
        passband_left_one,
        passband_left_two,
        passband_right_one,
        passband_right_two,
        nyq]

desired = [0, 0, 1, 1, 0, 0]

b = signal.firls(tap, band, desired, nyq=nyq)
w, h = signal.freqz(b)
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.title('Digital filter frequency response, N = ' + str(len(b)))
plt.plot(w * 30 / 2 /np.pi  , 20 * np.log10(abs(h)), 'b')
plt.ylabel('Amplitude [dB]', color='b')
plt.grid()
plt.axis('tight')

plt.subplot(2, 1, 2)
angles = np.unwrap(np.angle(h))
plt.plot(w / np.pi * 15 , angles, 'g')
plt.ylabel('Angle (radians)', color='g')
plt.grid()
plt.axis('tight')
plt.xlabel('Frequency [0 to Nyquist Hz, normalized]')
plt.figure()
plt.show()
```

```python
rawDict = {}
rawDict['data'] = scipy.io.loadmat('videoData.mat')
data = rawDict['data']['videodata']
```

```python
red_frame_average = np.zeros(300)
x = np.array(data)
for i in range(300):
    totalred = np.sum(x[:,:,0,i])
    red_frame_average[i] = totalred*1.0/(401*401)
    
filtered = signal.convolve(red_frame_average,b,mode = 'same')
plt.plot(red_frame_average)
plt.figure()
result = filtered[len(b):len(red_frame_average)-tap+1]

plt.plot(result)
plt.show()
```

```python
def peak_detection(t,sigs,dt=1,thresh=-100):
    peaks = []
    N = len(sigs)

    for i in range(dt,N-dt-1):
        left_neighbor = sigs[i-dt:i+1]
        right_neighbor = sigs[i:i+dt+1]
        if max(left_neighbor) == sigs[i] and max(right_neighbor) == sigs[i] and sigs[i] >= thresh:
            peaks.append((t[i],sigs[i]))

    return np.array(peaks)
```

```python
max_peaks = peak_detection(range(len(result)), result)
plt.plot(red_frame_average)
plt.figure()
plt.scatter(max_peaks[:,0], max_peaks[:,1], color = 'red')
plt.plot(result)
plt.show()
```
